# Bayesian Search Parameter Configuration and Input File Generation

This notebook is used to configure parameters for the Bayesian Search algorithm and generate the corresponding `.input` files.

### Main Features:
1. **Parameter Configuration**: Define the `dict_variables` dictionary to set search ranges or fixed values for the conformational free energies (G), free energy barriers of transitions (AG), and the binding free energies (BG).
2. **Experimental Data Integration**: Set experimental observation data (e.g., hydrolysis rates at different ATP concentrations).
3. **Model Initialization**: Call the `Input` class to configure the Markov model structure, active conformations, etc.
4. **Input File Generation**: Run the script to generate input configuration files for the subsequent C-language programs.
