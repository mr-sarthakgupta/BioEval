```python
confs = "ohca"
v_index = 323
activeConfs = "c"

param_indices = range(1, 501)
pred_folder="./example_predictions"
```

```python
variant = "%s-%d-%s"%(confs,v_index, activeConfs)

if activeConfs == "ca":
    activeConf_list = [2, 3]
else:
    activeConf_list = [2]
```

```python
plot_predictions(variant, pred_folder=pred_folder, param_indices=param_indices, outfilename="%s"%variant)
```

```python
cross_sets = []

for k in range(1, 501):
    
    flag = instance.load("%s/SteadyStates_cD=1.00e-09_set=%d.csv"%(pred_folder, k))
    if not flag:
        continue
            
    if (instance.data["pop_cat"][-1] - instance.data["pop_cat"][0] > 0.8)\
        and (instance.data["pop_wait"][6] > 0.2):
            cross_sets.append((k, instance.data["pop_cat"][-1] - instance.data["pop_cat"][0]))


sorted_sets = sorted(cross_sets, key=lambda x: x[1], reverse=True)
sets = [x[0] for x in cross_sets]


with open("cross_%s.csv"%variant, "w") as file:
    for k in sets:
        file.write("%d\n"%k)

data = np.loadtxt("cross_%s.csv"%variant, delimiter=",")
sets = list(np.array(data, dtype=int))

plot_population_shift(variant, pred_folder=pred_folder, param_indices=sets, outfilename="%s_cross"%variant)
plot_occ_ratio(variant, pred_folder=pred_folder, param_indices=sets, outfilename="%s_cross"%variant)

param_indices_sorted = \
CoarseGrainedPathway(shadowColors = ["palevioletred", "lightcyan", "#FFE9EB", "lightskyblue", "pink", "cornflowerblue"])\
    (["%s/SteadyStates_cD=1.00e-09_set=%d.csv"%(pred_folder, k) for k in sets],
     outFileName="pathway_%s_cross"%variant, tick=50)
```
