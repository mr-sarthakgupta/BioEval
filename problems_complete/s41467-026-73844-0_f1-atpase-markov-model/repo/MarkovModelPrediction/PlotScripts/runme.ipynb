```python
cross_sets = []

for k in range(1, 501):
    
    flag = instance.load("%s/SteadyStates_cD=1.00e-09_set=%d.csv"%(input_dir, k))
    if not flag:
        continue
            
    if (instance.data["pop_cat"][-1] - instance.data["pop_cat"][0] > 0.8)\
        and (instance.data["pop_wait"][6] > 0.2):
            cross_sets.append((k, instance.data["pop_cat"][-1] - instance.data["pop_cat"][0]))


sorted_sets = sorted(cross_sets, key=lambda x: x[1], reverse=True)
sets = [x[0] for x in cross_sets]
```

```python
plot_population_shift(pred_folder=input_dir, 
                      param_indices=sets, 
                      outfilename="%s_cross"%out_file_suffix,
                      workdir=work_dir)
```

```python
plot_occ_ratio(pred_folder=input_dir, 
               param_indices=sets, 
               outfilename="%s_cross"%out_file_suffix,
               workdir=work_dir)
```
